```python
# ==========================================================
# CS667 – Practical Data Science
# Project #2: Machine Learning (Regression Analysis)
# Step 1: Data Preparation (clean + future-proof)
# ==========================================================

import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# 1️⃣ Load dataset
df = pd.read_excel("/content/retail_data_synthetic_50k.xlsx")

# 2️⃣ Basic inspection
print("=== Dataset Overview ===")
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
print("\nData Types:\n", df.dtypes)

# 3️⃣ Drop duplicates & irrelevant columns
df.drop_duplicates(inplace=True)
df.drop(columns=['Transaction_ID', 'Customer_ID'], inplace=True, errors='ignore')

# 4️⃣ Handle missing values — warning-free modern syntax
num_cols = ['Age', 'Quantity', 'Unit_Price', 'Discount']
cat_cols = ['Gender', 'Category', 'Store_Region', 'Online_Or_Offline', 'Payment_Method']

for c in num_cols:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors='coerce')
        df[c] = df[c].fillna(df[c].median())

for c in cat_cols:
    if c in df.columns:
        df[c] = df[c].fillna('Unknown')

# 5️⃣ Convert string/currency to numeric
if 'Unit_Price' in df.columns:
    df['Unit_Price'] = pd.to_numeric(df['Unit_Price'].astype(str).str.replace('$',''), errors='coerce')
if 'Total_Amount' in df.columns:
    df['Total_Amount'] = pd.to_numeric(df['Total_Amount'].astype(str).str.replace('$',''), errors='coerce')

# 6️⃣ Feature engineering (temporal)
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['day_of_week'] = df['Date'].dt.dayofweek
df['month'] = df['Date'].dt.month
df['is_weekend'] = df['day_of_week'].isin([5,6]).astype(int)

def season_from_month(m):
    if m in [12, 1, 2]:  return 'Winter'
    if m in [3, 4, 5]:   return 'Spring'
    if m in [6, 7, 8]:   return 'Summer'
    return 'Fall'

df['season'] = df['month'].apply(season_from_month)

# 7️⃣ Define ColumnTransformer (One-Hot Encoding for categoricals)
categorical_features = ['Gender', 'Category', 'Store_Region',
                        'Online_Or_Offline', 'Payment_Method', 'season']

preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ],
    remainder='passthrough'  # keep numeric columns
)

# 8️⃣ Verify output
print("\n✅ Cleaned dataset summary")
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
display(df.head(5))


```

```python
# ==========================================================
# CS667 – Practical Data Science
# Project #2: Machine Learning (Regression Analysis)
# Step 2: Feature Selection & Data Splitting
# ==========================================================

from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd

# 1️⃣ Confirm dataset is loaded
print("Dataset shape before split:", df.shape)
print("Columns:", df.columns.tolist())

# ----------------------------------------------------------
# 2️⃣ Define target variable (dependent)
# ----------------------------------------------------------
target = 'Total_Amount'

# Check if target exists
if target not in df.columns:
    raise ValueError(f"Target column '{target}' not found. "
                     f"Available columns: {df.columns.tolist()}")

y = df[target]  # dependent variable

# ----------------------------------------------------------
# 3️⃣ Define predictor variables (independent features)
# ----------------------------------------------------------
# Drop non-predictive columns + target
drop_cols = ['Total_Amount', 'Date']  # Date already used for derived features
X = df.drop(columns=drop_cols, errors='ignore')

# Confirm final feature set
print("\nFeature columns used for modeling:\n", X.columns.tolist())
print("\nTarget variable:", target)

# ----------------------------------------------------------
# 4️⃣ Train-Test Split (80% train / 20% test)
# ----------------------------------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42
)

print("\n✅ Data split completed successfully!")
print("Training set size:", X_train.shape[0])
print("Testing set size:", X_test.shape[0])

# ----------------------------------------------------------
# 5️⃣ Optional: Preview samples
# ----------------------------------------------------------
print("\nSample of training data:")
display(X_train.head(5))
print("\nSample of target variable:")
display(y_train.head(5))

# ----------------------------------------------------------
# 6️⃣ Verify no missing values remain
# ----------------------------------------------------------
print("\nMissing values in training data:\n", X_train.isnull().sum())
```

```python
# ==========================================================
# CS667 – Practical Data Science
# Project #2: Machine Learning (Regression Analysis)
# Step 3: Model Building (with preprocessing)
# ==========================================================

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import numpy as np
import pandas as pd

# ----------------------------------------------------------
# 1️⃣ Identify categorical columns that need encoding
# ----------------------------------------------------------
categorical_cols = ['Gender', 'Category', 'Store_Region',
                    'Online_Or_Offline', 'Payment_Method', 'season']

numeric_cols = [c for c in X_train.columns if c not in categorical_cols]

# Preprocessor: OneHotEncode categoricals, pass numeric untouched
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
    ],
    remainder='passthrough'
)

# ----------------------------------------------------------
# 2️⃣ Define models
# ----------------------------------------------------------
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(random_state=42),
    "Random Forest": RandomForestRegressor(n_estimators=200, random_state=42, n_jobs=-1),
    "Gradient Boosting": GradientBoostingRegressor(random_state=42),
    "XGBoost": XGBRegressor(
        n_estimators=200,
        learning_rate=0.05,
        max_depth=6,
        random_state=42,
        n_jobs=-1,
        objective="reg:squarederror"
    )
}

# ----------------------------------------------------------
# 3️⃣ Train + evaluate each model through a Pipeline
# ----------------------------------------------------------
results = []

for name, model in models.items():
    # Create a full pipeline with preprocessing + model
    pipe = Pipeline([
        ('preprocess', preprocessor),
        ('model', model)
    ])

    # Train
    pipe.fit(X_train, y_train)

    # Predict
    y_pred = pipe.predict(X_test)

    # Metrics
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)

    results.append({
        "Model": name,
        "R²": round(r2, 4),
        "MAE": round(mae, 2),
        "MSE": round(mse, 2),
        "RMSE": round(rmse, 2)
    })

# ----------------------------------------------------------
# 4️⃣ Compare results
# ----------------------------------------------------------
results_df = pd.DataFrame(results).sort_values(by="R²", ascending=False).reset_index(drop=True)
print("\n=== Regression Model Performance Comparison ===")
display(results_df)

```

```python
# ==========================================================
# CS667 – Practical Data Science
# Project #2: Machine Learning (Regression Analysis)
# Step 4: Model Evaluation & Visualization
# ==========================================================

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# Re-train the best model (from Step 3)
best_model_name = "Random Forest"   # or "XGBoost", based on your Step 3 results
best_model = models[best_model_name]

# Build full pipeline (preprocessing + best model)
pipe = Pipeline([
    ('preprocess', preprocessor),
    ('model', best_model)
])

# Fit and predict again for evaluation visuals
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)

# ----------------------------------------------------------
# 1️⃣ Compute regression metrics
# ----------------------------------------------------------
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("=== Regression Model Evaluation Metrics ===")
print(f"R² (Coefficient of Determination): {r2:.4f}")
print(f"MAE (Mean Absolute Error): {mae:.2f}")
print(f"MSE (Mean Squared Error): {mse:.2f}")
print(f"RMSE (Root Mean Squared Error): {rmse:.2f}")

# ----------------------------------------------------------
# 2️⃣ Actual vs Predicted Plot
# ----------------------------------------------------------
plt.figure(figsize=(7,6))
sns.scatterplot(x=y_test, y=y_pred, alpha=0.6, color='navy')
plt.xlabel("Actual Total Amount")
plt.ylabel("Predicted Total Amount")
plt.title(f"Actual vs Predicted Values ({best_model_name})")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', lw=2)
plt.grid(True)
plt.show()

# ----------------------------------------------------------
# 3️⃣ Residual Plot
# ----------------------------------------------------------
residuals = y_test - y_pred
plt.figure(figsize=(7,5))
sns.scatterplot(x=y_pred, y=residuals, color='teal', alpha=0.6)
plt.axhline(0, color='red', linestyle='--', lw=2)
plt.xlabel("Predicted Values")
plt.ylabel("Residuals (Actual - Predicted)")
plt.title(f"Residual Plot ({best_model_name})")
plt.grid(True)
plt.show()

# ----------------------------------------------------------
# 4️⃣ Feature Importance (for tree-based models)
# ----------------------------------------------------------
if best_model_name in ["Random Forest", "Gradient Boosting", "XGBoost"]:
    # Extract encoded feature names
    encoded_features = preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_cols)
    all_features = np.concatenate([encoded_features, numeric_cols])

    importances = best_model.feature_importances_
    feat_imp = pd.DataFrame({'Feature': all_features, 'Importance': importances})
    feat_imp = feat_imp.sort_values(by='Importance', ascending=False).head(15)

    plt.figure(figsize=(8,5))
    sns.barplot(x='Importance', y='Feature', data=feat_imp, palette='viridis')
    plt.title(f"Top 15 Feature Importances – {best_model_name}")
    plt.xlabel("Relative Importance")
    plt.ylabel("Feature Name")
    plt.show()

# ----------------------------------------------------------
# 5️⃣ Display summary table
# ----------------------------------------------------------
summary_df = pd.DataFrame({
    'Metric': ['R²', 'MAE', 'MSE', 'RMSE'],
    'Value': [r2, mae, mse, rmse]
})
display(summary_df)
```
